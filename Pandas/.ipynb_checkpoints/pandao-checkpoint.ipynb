```python
import pandas as pd
import faker as fk

fake = fk.Faker(["En_In", "Gu_In"])

name_list = [fake.name() for i in range(10)]
s_names = pd.Series(data = name_list, name = "Name", index = range(1001, 1011))
s_names
```

```python
email_list = [fake.email(safe = True, domain = "gmail.com") for i in range(10)]

s_emails = pd.Series(data = email_list, name = "Email", index = range(1001, 1011))
s_emails
```

```python
s_date = pd.Series(
                    data = 
                    [
                        fake.date(
                            pattern = '%d-%m-%y',
                            end_datetime = pd.to_datetime('01-01-2000'))
                        for i in range(10)
                    ],
                    index = range(1001, 1011)
                    )
s_date
```

```python
df = pd.DataFrame(
                data = {
                    'Name' : s_names,
                    'Email' : s_emails,
                    'DOB' : s_date
                },
                index = s_names.index
)

df
```

```python
df.info()
```

```python
df['DOB'] = pd.to_datetime(df['DOB'])
df.info()
```

```python
# create a df with f_name, l_name, email, contact

import pandas as pd
import faker as fk

fake = fk.Faker(['En_In'])
```

```python
f_name_list = [
    fake.first_name() for i in range(100)
]

l_name_list = [
    fake.last_name() for i in range(100)
]

email_list = [
    fake.email(domain = "gmail.com")
    for i in range(100)
]

contact_list = [
    fake.phone_number() for i in range(100)
]
```

```python
s_fname = pd.Series(data = f_name_list)
s_lname = pd.Series(data = l_name_list)
s_email = pd.Series(data = email_list)
s_contact = pd.Series(data = contact_list)
```

```python
df = pd.DataFrame(
    {
        'First Name' : s_fname,
        'Last Name' : s_lname,
        'Email ID' : s_email,
        'Contact' : s_contact
    }
)

df
```
