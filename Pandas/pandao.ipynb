```python
import numpy as np  # Import NumPy for numerical operations
import pandas as pd  # Import pandas for data manipulation

# Define a dictionary containing student names and their scores in various subjects
data = {
    'Name': ['rohit', 'aakash', 'yash', 'piyush', 'ankit', 'ashutosh', 'kamlesh', 'himanshu'],
    'Maths 1': np.random.randint(40, 100, 8),            # Random scores from 40 to 99 for 8 students
    'Maths 2': np.random.randint(40, 100, 8),
    'Physics': np.random.randint(40, 100, 8),
    'Chemistry 1': np.random.randint(40, 100, 8),
    'Chemistry 2': np.random.randint(40, 100, 8),
    'English': np.random.randint(40, 100, 8),
    'Computer Science 1': np.random.randint(40, 100, 8),
    'Computer Science 2': np.random.randint(40, 100, 8),
    'Environmental Studies': np.random.randint(40, 100, 8),
}

# Convert the dictionary to a pandas DataFrame, creating columns for each key
df = pd.DataFrame(data=data)
```

```python
# Select and display the names of students who scored more than 60 in Physics
df[df['Physics'] > 60]['Name']
```

```python
# Select rows where the Physics score is greater than 40 and less than 60
df[(df['Physics'] > 40) & (df['Physics'] < 60)]
```

```python
# Select rows where the Physics score is greater than 40 and less than 60
# using chained indexing, not recommended may lead to errors.
df[(df['Physics'] > 40)][(df['Physics'] < 60)]
```

## FAKER

```python
import pandas as pd  # Import pandas for data handling
import faker as fk   # Import Faker library for generating fake data

# Initialize Faker with English (India) and Gujarati (India) locales
fake = fk.Faker(["En_In", "Gu_In"])

# Generate a list of 10 fake names
name_list = [fake.name() for i in range(10)]

# Create a pandas Series from the generated names
# - Data: name_list  
# - Series name: "Name"  
# - Custom index: integers from 1001 to 1010
s_names = pd.Series(data=name_list, name="Name", index=range(1001, 1011))

# Display the Series
s_names
```

```python
# Generate a list of 10 fake, safe email addresses with the domain "gmail.com"
email_list = [fake.email(safe=True, domain="gmail.com") for i in range(10)]

# Create a pandas Series from the generated email addresses
# - Data: email_list
# - Series name: "Email"
# - Custom index: integers from 1001 to 1010
s_emails = pd.Series(data=email_list, name="Email", index=range(1001, 1011))

# Display the Series
s_emails
```

```python
import pandas as pd  # Import pandas library

# Generate a pandas Series of 10 fake dates (strings) in 'dd-mm-yy' format
# Dates are randomly generated before '01-01-2000'
s_date = pd.Series(
    data=[
        fake.date(
            pattern='%d-%m-%y',
            end_datetime=pd.to_datetime('01-01-2000')
        )
        for i in range(10)
    ],
    index=range(1001, 1011)
)

# Display the Series
s_date
```

```python
import pandas as pd  # Import pandas for data manipulation

# Create a DataFrame from three pandas Series: names, emails, and dates of birth (DOB)
df = pd.DataFrame(
    data={
        'Name': s_names,    # Series of fake names with index 1001-1010
        'Email': s_emails,  # Series of fake emails with same index
        'DOB': s_date       # Series of fake dates of birth with same index
    },
    index=s_names.index     # Use the same custom index as the 'Name' Series
)

# Display the DataFrame
df
```

```python
df.info()
```

```python
# Convert the 'DOB' column in df from string format to pandas datetime objects
df['DOB'] = pd.to_datetime(df['DOB'])

# Display concise summary of the DataFrame including data types, non-null counts, and memory usage
df.info()
```

```python
import pandas as pd  # Import pandas for data manipulation
import faker as fk   # Import Faker library for generating fake data

# Initialize Faker with English (India) locale
fake = fk.Faker(['En_In'])

# Generate 100 fake first names
f_name_list = [fake.first_name() for i in range(100)]

# Generate 100 fake last names
l_name_list = [fake.last_name() for i in range(100)]

# Generate 100 fake email addresses with gmail.com domain
email_list = [fake.email(domain="gmail.com") for i in range(100)]

# Generate 100 fake phone numbers
contact_list = [fake.phone_number() for i in range(100)]

# Convert each list into a pandas Series
s_fname = pd.Series(data=f_name_list)
s_lname = pd.Series(data=l_name_list)
s_email = pd.Series(data=email_list)
s_contact = pd.Series(data=contact_list)

# Create a DataFrame by combining the Series as columns
df = pd.DataFrame(
    {
        'First Name': s_fname,
        'Last Name': s_lname,
        'Email ID': s_email,
        'Contact': s_contact
    }
)

# Display the resulting DataFrame
df
```
