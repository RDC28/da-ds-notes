```python
import pandas as pd  # Import pandas for data manipulation

# Read the Titanic dataset from a CSV file into a pandas DataFrame
df = pd.read_csv("../Datasets/titanic.csv")

# Display the DataFrame (first and last 5 rows by default)
df
```

```python
df.head()
```

```python
df.info()
```

```python
# Select rows where the passenger did not survive (Survived == 0) and is male (Sex == 'male')
df[(df['Survived'] == 0) & (df['Sex'] == 'male')]
```

```python
# Select rows where the 'Age' value is missing (NaN)
df[df['Age'].isnull()]
```

```python
# Get the shape (rows, columns) of the subset of DataFrame where 'Age' is missing
df[df['Age'].isnull()].shape
```

```python
# Select rows 0 to 9 (10 rows) and columns 3 to 5 (3 columns) from the DataFrame using integer-location based indexing
df.iloc[0:10, 3:6]
```

```python
# Select specific rows with integer positions 1, 3, and 87
# and columns at positions 3, 4, and 5 from the DataFrame
df.iloc[[1, 3, 87], 3:6]
```

```python
# Use .iloc to select rows with null 'Age' values, and all columns

# Step 1: df['Age'].isnull() creates a boolean mask for null 'Age' rows
# Step 2: df[df['Age'].isnull()].index gets the integer positions (indexes) of these rows
# Step 3: df.iloc[...] selects rows by integer positions, ':' means all columns

df.iloc[df[df['Age'].isnull()].index, :]
```

```python
# Retrieve the Index object containing all column names of the DataFrame
df.columns
```

```python
# Select rows where 'Survived' is 1 (passengers who survived)
# and show only the 'Survived' and 'Name' columns using label-based indexing
df.loc[df[df['Survived'] == 1].index, ["Survived", 'Name']]
```

```python
# Retrieve all unique values present in the 'Age' column
df['Age'].unique()
```

```python
# Count the occurrences of each unique value in the 'Pclass' column
df['Pclass'].value_counts()
```

```python
# Create a DataFrame from the value counts of 'Pclass'
pd.DataFrame(df['Pclass'].value_counts())
```

```python
# Create a DataFrame from the value counts of 'Pclass' and retrieve its index
pd.DataFrame(df['Pclass'].value_counts()).index
```

```python
# Calculate the number of missing (null) values in each column of the DataFrame
df.isnull().sum()
```

```python
# Display the data types of each column in the DataFrame
df.dtypes
```

```python
# Drop rows with index labels 1, 2, and 4 from the DataFrame and return a new DataFrame
df.drop([1, 2, 4])
```

```python
# Remove rows with index labels 412 and 379 from the DataFrame, modifying df directly
df.drop([412, 379], inplace=True)

# Display the updated DataFrame
df
```

```python
# Remove the column 'Parch' from the DataFrame, modifying df in place
df.drop(['Parch'], axis=1, inplace=True)

# Display the updated DataFrame
df
```
