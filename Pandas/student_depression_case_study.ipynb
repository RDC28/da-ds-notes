```python
import numpy as np
import pandas as pd
import datetime
```

```python
df = pd.read_csv('../Datasets/Student Depression Dataset.csv')
df
```

```python
# df.insert(2, 'Birth Year', np.nan)
df
```

```python
# calculate birth year
curr_year = datetime.datetime.now().year
df['Birth Year'] = curr_year - df['Age']
df
```

```python
# converting 'yes' and 'no' values of family history to 1 and 0.

df.loc[df['Family History of Mental Illness'] == 'Yes', 'Family History of Mental Illness'] = 1
df.loc[df['Family History of Mental Illness'] == 'No', 'Family History of Mental Illness'] = 0
df
```

```python
# removing the last 3 unnamed columns.

df.drop(['Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23'], axis = 1, inplace = True)
```

```python
df
```

```python
df.drop_duplicates(inplace = True)
```

```python
# remove entries of people whose profession is other than stuent.

df.drop(df.loc[df['Profession'] != 'Student', 'Profession'].index, inplace = True)
```

```python
# reset index
df.reset_index(drop = True, inplace = True)
df
```

```python
# remove column work pressure
df.drop('Work Pressure', axis = 1, inplace = True)
```

```python
# round off the cgpa into 2 digit after decimal point
df['CGPA'] = df['CGPA'].round(2)
df
```

```python
# remove records with 0 cgpa2
df.drop(df.loc[df['CGPA2'] == 0, 'CGPA2'].index, inplace = True)
df.reset_index(drop = True, inplace = True)
```

```python
# rename column Work/Study Hours to study hours.
df.rename(columns = {'Work/Study Hours': 'Study Hours'}, inplace = True)
```

```python
# remove id column
df.drop('id', axis = 1, inplace = True)
df.reset_index(drop = True, inplace = True)
```

```python
# filter records


# replace 'L' and 'M'
df.loc[df['Sleep Duration2'] == 'L'] = pd.Series(np.arange(0, 5)).mean().astype(np.int64)
df.loc[df['Sleep Duration2'] == 'M'] = pd.Series(np.arange(8, 12)).mean().astype(np.int64)

# convert sleep duration2 to float
df['Sleep Duration2'] = pd.to_numeric(df['Sleep Duration2'], errors='coerce')

# actual query
df[(df['Sleep Duration2'] > df['Sleep Duration2'].mean()) & 
    (df['Dietary Habits'] == 'Healthy') &
    (df['Suicidal thought'] == 1)
    ]
```
